```python
#Neuron
x_1 = 2  #These x_i's are input to the neuron.
x_2 = 4
x_3 = 6
w_1 = .1 #These w_i's are the weights to the given neuron.
w_2 = .2
w_3 = .3
b = -2     #b is the bias (the cut off point on which the activation function make decisions.)
#Computing the result
# y= (x_i*w_i +b)
y = (x_1*w_1 + x_2*w_2 + x_3*w_3 + b)
if y > 0:
    print("yes")
else:
    print("no")
```
